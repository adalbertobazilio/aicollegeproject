```python
import time
from google import genai
from google.genai import types
from google.colab import userdata

# 1. Autenticação
API_KEY = userdata.get('API_KEY')
client = genai.Client(api_key=API_KEY)

# 2. Definição do Robô (Personalidade + Conhecimento + Regras)
# Aqui é onde você ganha os pontos de "não alucinar"
instrucoes_sistema = """
Você é o Especialista de Suporte da "TechFlow Solar".
Seu conhecimento é restrito ao Inversor TF-X100.

CONHECIMENTO PRIVADO:
- Erro E01: Sobreaquecimento. Solução: Limpar as saídas de ar laterais.
- Erro E02: Falha de conexão. Solução: Verificar o cabo azul de dados.
- Garantia: 5 anos, mas perde a validade se o selo de papel interno for rompido.

REGRAS RÍGIDAS DE FLUXO:
1. Responda APENAS sobre os erros E01, E02 e Garantia.
2. Se o usuário perguntar algo fora disso (ex: futebol, clima, outros produtos), diga: "Sinto muito, sou um especialista restrito ao modelo TF-X100 e não possuo essa informação."
3. Você deve permitir exatamente 3 interações.
4. Na 3ª (última) resposta, você deve obrigatoriamente:
   - Responder a dúvida técnica.
   - Fazer um resumo curto do que foi tratado na conversa.
   - Finalizar com a frase: "ATENDIMENTO ENCERRADO."
"""

# 3. Inicializando o chat
chat = client.chats.create(
    model="gemini-2.5-flash", # O 1.5 Flash é o mais estável para cota gratuita
    config=types.GenerateContentConfig(
        system_instruction=instrucoes_sistema,
        temperature=0.0 # Temperatura zero evita que ele "invente" coisas
    )
)

print("--- Suporte TechFlow Solar Ativado ---")
print("(Você tem direito a 3 perguntas)\n")

for i in range(1, 4):
    pergunta = input(f"Pergunta {i}/3: ")

    # Pequeno delay de 2 segundos para evitar erro de cota por pressa
    time.sleep(2)

    # Envia a pergunta. Na última, reforçamos a ordem de encerrar.
    prompt_final = pergunta
    if i == 3:
        prompt_final += " (Lembre-se: esta é a última, faça o resumo e encerre o chat)."

    try:
        response = chat.send_message(prompt_final)
        print(f"\nBot: {response.text}\n")
        print("-" * 50)
    except Exception as e:
        print(f"\nErro de conexão: {e}. Tente aguardar 30 segundos.")
        break

print("\nSessão finalizada. Pronto para gravar o vídeo!")
```
